## Connect to SAP HANA Cloud and check successful connection

Check Python Machine Learning client for SAP HANA version (aka hana-ml) installed and print.

```python
## Loading the Python Machine Learning client library for SAP HANA and get the version
import hana_ml
print(f'Python Machine Learning client for SAP HANA: {__import__("hana_ml").__version__}')
```

Load the user connection data from a file for automatically establishing the connection using the prepared credentials

üëâüèª The content of the [./user.ini](./images/BAS_python_edit_userini.png)-file will be processed in the next step.


![Set user](./images/BAS_python_edit_userini.png)

```python
import os, sys

file_path = '../ex0/user.ini'  

if os.path.exists(file_path):
    print(f"The file '{file_path}' exists.")
    from hana_ml.algorithms.pal.utility import Settings
    myhost, myport, myuser, mypwd = Settings.load_config(file_path)
    if myhost == '' or myport == 0 or myuser == '' or mypwd == '':
        print("At least one required connection parameter is empty.")
    else:
        print(f"Connection to SAP HANA database will be attempted for the user {myuser} at SQL endpoint {myhost}:{myport} ")
else:
    sys.exit(f"The file '{file_path}' does not exist! Create and configure it first.")
    
```

```python
from hana_ml import dataframe as hdf
myconn=hdf.ConnectionContext(
    address=myhost, 
    port=myport, 
    user=myuser,
    password=mypwd
)
# Control connection
print('Connection to SAP HANA Cloud successful is: ', myconn.connection.isconnected(), ' , with database version: ', myconn.hana_version())
print(f"Connected to SAP HANA Cloud database at {myhost}:{myport} as user {myuser}")
```

Execute an SQL query on SAP HANA database and get the result displayed

```python
print(f'Current time on the SAP HANA Cloud database service: {myconn.sql("SELECT NOW() FROM DUMMY").collect().CURRENT_TIMESTAMP[0]}')
```
